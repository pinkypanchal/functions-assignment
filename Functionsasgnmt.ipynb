**1. What is the difference between a function and a method in Python?**
  - **Function:** A function is a block of code that performs a specific task and can be called independently. It is defined using the **def** keyword.
  - eg:

```
def add(a, b):
    return a + b
```



 - **Method:** A method is a function that is associated with an object and is called on that object. It is defined within a class.
 - eg:

```
class Calculator:
    def add(self, a, b):
        return a + b

calc = Calculator()
result = calc.add(5, 3)  # Calling a method
```

**6. Explain the concept of generators in Python and how they are defined.**
  - Generators are a type of iterable that generate values on the fly and use the yield statement to produce a series of values. They are defined like regular functions but use yield instead of return.
  - eg:



```
def count_up_to(n):
    count = 1
    while count <= n:
        yield count
        count += 1

for number in count_up_to(5):
    print(number)  # Outputs 1, 2, 3, 4, 5
```



**7. What are the advantages of using generators over regular functions?**
  
* Memory efficient as values are generated one at a time instead of storing them all at once.  
- Faster execution due to on-demand computation without unnecessary memory allocation.  
- Lazy evaluation allows handling large datasets efficiently without loading everything into memory.  
- No need for explicit storage of large lists, reducing overhead and improving performance.  
- Can generate infinite sequences like Fibonacci without excessive memory consumption.  
- Simplifies code by reducing the need for temporary lists or complex loop structures.  
- Retains state between yield calls, avoiding redundant computations.  
- Works efficiently in data pipelines with  map(), filter(), and reduce().

**10. What is the difference between map(), reduce(), and filter() functions in Python?**
 - map(): Applies a function to all items in an iterable.

```
nums = [1, 2, 3, 4]
squared = map(lambda x: x**2, nums)
print(list(squared))  # Output: [1, 4, 9, 16]
```

 - filter(): Filters items in an iterable based on a condition.

```
nums = [1, 2, 3, 4]
even = filter(lambda x: x % 2 == 0, nums)
print(list(even))  # Output: [2, 4]
```
 - reduce(): Applies a function cumulatively to the items of an iterable, reducing it to a single value.

```
from functools import reduce
nums = [1, 2, 3, 4]
sum = reduce(lambda x, y: x + y, nums)
print(sum)  # Output: 10
```

**4. Write a Python function that checks if a given number is prime or not from 1 to 200.**

````python
def prime_or_not(num):
    '''
    this function is created to check if a number is prime or not
    arguments:
    num = int(numbers that will be checked)
    return:
    true if the number is prime otherwise false
    '''
    if num <= 1:
        return False #numbers less than or equal to 1 are not prime
    for i in range (2,int(num*0.5) + 1 ):
        if num%i == 0:
            return False #number is not prime
    return True #number is prime

for number in range(1,201):
    if prime_or_not(number):
        print(number," is a prime number")
````

**5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.**

````python
class FibonacciIterator:
    def __init__(self, terms):
        self.terms = terms
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < self.terms:
            result = self.a
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return result
        else:
            raise StopIteration

# Example usage
fibonacci = FibonacciIterator(10)
for num in fibonacci:
    print(num, end=' ')
````

**6. Write a generator function in Python that yields the powers of 2 up to a given exponent.**

````python
def powers_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

# Example usage
for power in powers_of_two(10):
    print(power, end=' ')
````

**7. Implement a generator function that reads a file line by line and yields each line as a string.**

````python
def read_line_by_line(file_path):
    '''
    this function read file line by line and yields each line as a string
    parameters:
    file_path is the path of the file which function will read line by line and will return each line as a string
    '''
    try:
        with open(file_path,'r') as file:
            for line in file:
                yield line.strip()
    except FileNotFoundError:
        print(f"Error, File {file_path} was not found")
    except IOError:
        print(f"Error: an error occured while reading the file {file_path}")

file_path = "example.txt"
print("File content line by line")

for file in read_line_by_line(file_path):
    print(file)
````

**10. Create a Python program that uses filter() to remove all the vowels from a given string.**

````python
def remove_vowels(s):
    vowels = 'aeiouAEIOU'
    return ''.join(filter(lambda x: x not in vowels, s))

# Example usage
input_string = "Hello, World!"
result_string = remove_vowels(input_string)
print(result_string)
````
